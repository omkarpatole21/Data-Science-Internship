```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
x = np.array([12,49,34,23,87,46,30])

y = np.array([10,23,31,58,99,37,49])

z = np.array([98,76,67,89,50,55,27])
```

## Line Plot

```python
plt.plot(x,y, marker='o', color='blue', mfc='r' , ls='--', ms=7)
plt.grid()
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')
plt.show()
# marker => <,>,^,o,*,H
```

```python
plt.plot(x,z, marker='o', color='blue', mfc='r' , ls='--', ms=7)
plt.grid()
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')
plt.show()
```

```python
plt.plot(x,y, marker='o', color='blue', mfc='r' , ls='--', ms=7)
plt.plot(x,z, marker='o', color='orange', mfc='r' , ls='--', ms=7)
plt.grid()
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')
plt.plot(x,z, marker='*', color='cyan', mfc='blue' , ls='--', ms=7)
plt.grid()
plt.show()

plt.figure(figsize=(10,10))
plt.subplot(2,2,2)
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')
plt.plot(x,y, marker='o', color='cyan', mfc='blue' , ls='--', ms=7)
plt.grid()
plt.show()
```

## Bar Graph

```python
student = [56,65,46,51,61]

courses = ['Java', 'Python', 'Dart', 'R', 'Kotlin']
```

```python
plt.bar(courses, student, color='yellow', edgecolor='blue', width=0.8, linewidth=2, alpha=0.7);
plt.title('Bar plot')
plt.xlabel('Courses')
plt.ylabel('Students')
```

## Scatter Plot

```python
plt.scatter(x,y, marker='o');
plt.grid()
```

## Pie Chart

```python
plt.figure(figsize=(7,8))
plt.title('Pie Chart')
plt.pie(student, labels = courses, colors=['red','orange','yellow','brown','gray'], autopct='%.2f%%',shadow=True, radius=0.8, labeldistance=1.2, startangle=30,
       explode=[0,0,0.2,0.1,0]);
```

```python
plt.figure(figsize=(10,10))
plt.subplot(2,2,1)
plt.bar(courses, student, color='yellow', edgecolor='blue', width=0.8, linewidth=2, alpha=0.7);
plt.title('Bar plot')
plt.xlabel('Courses')
plt.ylabel('Students')

plt.subplot(2,2,2)
plt.plot(x,y, marker='o', color='blue', mfc='r' , ls='--', ms=7)
plt.grid()
plt.title('Line Plot')
plt.xlabel('first value')
plt.ylabel('Second value')

plt.subplot(2,2,3)
plt.title('Pie Chart')
plt.pie(student, labels = courses, colors=['red','orange','yellow','brown','gray'], autopct='%.2f%%',shadow=True, radius=0.8, labeldistance=1.2, startangle=30,);

plt.subplot(2,2,4)
plt.title('Scatter plot')
plt.scatter(x,y, marker='o');
plt.grid()

plt.savefig('my plot')
```

## Box Plot

```python
marks = [67,63,70,75,72,56,51,21,95]
```

```python
plt.boxplot(marks);
```

```python
q3=np.percentile(marks, 75)

q1=np.percentile(marks, 25)
```

```python
q3, q1
```

```python
iqr = q3-q1
iqr
```

```python
# lower bound
lb = q1 - 1.5 * iqr

# upper bound
up = q3 + 1.5 * iqr
```

```python
lb, up
```
